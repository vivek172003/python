## Seaborn:

- Data visualization library built on top of matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics.
- It simplifies the process of creating complex visualizations and integrates well with pandas data structures.
- Seaborn is especially suited for visualizing categorical and numerical data.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Seaborn provides built-in datasets that can be loaded for practice
tips = sns.load_dataset('tips')
tips
```

### Plot Types:-

Seaborn offers different types of plots for visualizing relationships, distributions, and categorical data. These plots are highly customizable and can be combined.

### 1. Relational Plots:

Relational plots show relationships between multiple variables, especially between numerical variables.

##### 1.1 Scatter Plot

A scatter plot shows relationships between two numeric variables or continuous features.

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

##### on how to change figure size (not part of scatter plot alone)

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))  # width = 10 inches, height = 6 inches
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
#hue - Grouping variable that will produce points with different colors
sns.scatterplot(data=tips, x="total_bill", y="tip", hue='sex', palette='Dark2')  # for categorical column, distinctive colors
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue='size', palette='rocket_r') # for continuous column, color gradient
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", size='tip') # based on tip value, bigger size point for higher value
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", s=70, alpha = 0.5)
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", style='time') # based on time value, points style changes
plt.savefig('myscatterplot.jpg', bbox_inches='tight') #bbox_inches - tight :This ensures that the saved plot won't have extra white space around the figure.
plt.show()
```

```python
tips['tip_rate'] = tips.eval("tip / total_bill") #arithmetic operations and conditional logic using eval, 
                                                 #same as tips["tip"] / tips["total_bill"]

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="tip_rate")

plt.show()
```

##### 1.2 Line Plot
Line plots are used for continuous data and time series.

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
may_flights = flights.query("month == 'May'") # same as - flights[flights['month'] == 'May']
sns.lineplot(data=may_flights, x="year", y="passengers")
plt.show()
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.legend(loc='upper left')
plt.show()
```

Common loc values for legend positioning:
- 'best': Automatically chooses the best position (default).
- 'upper left': Top-left corner.
- 'upper right': Top-right corner.
- 'lower left': Bottom-left corner.
- 'lower right': Bottom-right corner.
- 'center left': Centered vertically on the left.
- 'center right': Centered vertically on the right.
- 'center': Center of the plot.

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.legend(bbox_to_anchor=(1.3,1))  #(0, 0) is the bottom-left corner and (1, 1) is the top-right corner of the axes.
plt.show()
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month", style="month")
plt.show()
```

### 2. Distribution Plots:

Displays single continuous feature and helps visualize propoerties such as deviation or average values

##### 2.1 Histogram with histplot()
Histograms are used to display the distribution of numerical data by grouping into bins.

```python
tips.head()
```

```python
sns.histplot(data=tips, x="total_bill")
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", bins=100)
plt.show()
```

```python
#Kernel Density Estimate (KDE) - smooth approximation of the underlying distribution by using a Gaussian (normal) kernel
#to estimate the density of data points, providing a clearer sense of the data's overall shape.
sns.histplot(data=tips, x="total_bill", kde=True)
plt.show()
```

```python
sns.histplot(data=tips, x="total_bill", hue="sex", multiple="stack") #Histogram with Multiple Categories
plt.show()
```

##### 2.2 KDE Plot with kdeplot()
A KDE plot estimates the probability density function of a continuous variable.

```python
sns.kdeplot(data=tips, x="total_bill", fill=True)
plt.show()
```

```python
#KDE Plot for Multiple Variables
sns.kdeplot(data=tips, x="total_bill", hue="sex", fill=True)
plt.show()
```

```python
"""
bandwidth adjustment -  refers to the width of the kernel used to calculate the density estimate. 
A larger bandwidth results in a smoother curve, while a smaller bandwidth can capture more detail
but may also lead to a noisier appearance.
"""

sns.kdeplot(data=tips, x="total_bill", bw_adjust=0.3, color='red', fill=True)
plt.show()
```

##### 2.3. rugplot()
adds a rug plot to the existing axes, displaying individual data points along the x-axis. It's useful for visualizing the distribution of data points without the aggregation seen in histograms or KDE plots.

```python
sns.rugplot(data=tips, x="total_bill", color='blue')
plt.show()
```

##### 2.4. displot()
combines histograms, kernel density estimates (KDE) and rugplot in a flexible way, allowing for easy visualization of distributions. It can create faceted plots for different subsets of data.

```python
sns.set_style('darkgrid')
sns.displot(data=tips, x="total_bill", kde=True, bins=30, rug=True)
plt.show()
```

### 3. Categorical Plots
to display the relationship between categorical variables and numerical variables(equivalent to groupby call)

##### 3.1. catplot()
A high-level interface for drawing categorical plots. It can create various types of plots (box plots, bar plots, violin plots, etc.) based on the parameters you provide.

```python
# Create a categorical plot (bar plot in this case) 
# Vertical Line(error bar)-  shows the range of uncertainty/variability around that average . 
# Lower bound shows how low the average could be with some uncertainty, same contrary works for upper bound.

sns.catplot(data=tips, x="day", y="total_bill", kind="bar")

# Show the plot
plt.show()
```

```python
# Example with catplot and specifying size
# height=6 sets the height to 6 inches. aspect=1.5 makes the width 1.5 times the height, so the figure size will be 6x9 inches.

sns.catplot(data=tips, x="day", y="total_bill", kind="bar", height=6, aspect=1.5)
plt.show()
```

##### 3.4. boxplot()
Box plots display the distribution of numerical data and show the median, quartiles, and outliers.

Visual Description of a Boxplot
- Box: A rectangular box that extends from Q1 (the first quartile) to Q3 (the third quartile), containing the middle 50% of the data.
- Median Line: A horizontal line inside the box indicates the median (Q2) of the dataset.
- Whiskers: Lines extending from the box to the minimum and maximum data points within 1.5 times the IQR.
- Outliers: Points outside the whiskers, often shown as small dots or asterisks, represent data points that are considered outliers.

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
plt.show()
```

### 4.Comparison plots

 used to compare different categories or groups within a dataset. These plots help visualize differences in central tendency, variability, or distribution among categorical variables.

##### 4.1. jointplot()
visualize the relationship between two variables along with their individual distributions. It creates a scatter plot (or other types) at the center and histograms or kernel density estimates (KDE) on the sides.

```python
# Create a joint plot to visualize the relationship between total_bill and tip
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter")
plt.show()
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="scatter", hue='sex')
plt.show()
```

##### 4.2. pairplot()
creates a matrix of scatter plots for all possible pairs of numerical columns in a dataset. It helps to visualize relationships between multiple variables and can include histograms or KDEs on the diagonal.

```python
sns.pairplot(tips, hue="time", palette='viridis')
plt.show()
```

```python
sns.pairplot(tips, hue="time", diag_kind='hist')
plt.show()
```

```python
sns.pairplot(tips, hue="time", diag_kind='hist', corner=True) #removes duplicate scatter plots and keeps only 1 corner
plt.show()
```

### 5.Matrix plots

- Grid Format: The plot consists of rows and columns, creating a matrix layout. Each row and column corresponds to a specific variable.

- Color Representation: Each cell in the matrix is usually filled with color or a gradient, indicating the value of the relationship between the variables represented by that row and column. Darker or lighter colors may represent higher or lower values, respectively.

- Visual Comparison: Matrix plots are useful for comparing multiple variables at once, allowing you to quickly identify patterns, correlations, or anomalies in the data.

##### 5.1. heatmap()
graphical representation of data where individual values are represented as colors. Plot rectangular data as a color-encoded matrix.

```python
# Select only numeric columns
numeric_tips = tips.select_dtypes(include='number')

# Create the heatmap
sns.heatmap(numeric_tips.corr(), annot=True, cmap="magma", fmt=".2f")
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```

- 0.0 to 0.1: No or very weak correlation.
- 0.1 to 0.3: Weak positive correlation.
- 0.3 to 0.5: Moderate positive correlation.
- 0.5 to 0.7: Moderate to strong positive correlation.
- 0.7 to 0.9: Strong positive correlation.
- 0.9 to 1.0: Very strong positive correlation.

##### 5.2. clustermap()
type of heatmap that clusters rows and/or columns based on similarity, making it easier to identify patterns. It uses hierarchical clustering algorithms to group similar observations.

```python
# Create a cluster map

sns.clustermap(numeric_tips.corr(), annot=True, cmap="magma", fmt=".2f")
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```

### 6. Regression plot

relationship between two variables and the trend or pattern in that relationship.
- Scatter Plot with a Line: It starts with a scatter plot, where individual data points represent the relationship between two variables (e.g., height vs. weight). Then, a regression line is drawn to show the general trend or direction of the data.

- Line of Best Fit: The regression line represents the "best fit" through the data points, showing the average relationship between the two variables. The line helps predict one variable based on the other.

- Used for Predictions: Regression plots are helpful for making predictions. For example, you can predict the value of a variable (like weight) based on another variable (like height).

- Positive or Negative Trend: The line can have a positive slope (indicating that as one variable increases, the other also increases) or a negative slope (indicating that as one variable increases, the other decreases).

##### 6.1. regplot()
creates a basic regression plot between two variables
- Regression Line: A linear regression line is drawn that best fits the data.
- Confidence Interval (CI): By default, it includes a shaded confidence interval around the regression line, which shows the uncertainty of the regression estimate. - set at 95%, meaning that if we repeated the experiment many times, 95% of the time the true regression line would fall within the shaded area.
- Scatterplot: It automatically plots a scatterplot of the individual data points, over which the regression line is drawn.

```python
# Create a regression plot
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot: Total Bill vs Tip")
plt.show()
```

##### 6.2. lmplot()
 higher-level function that allows you to create regression plots with more customization, including faceting (splitting the data by categorical variables).

```python
# Create a regression plot with faceting by time (lunch vs dinner)
sns.lmplot(x="total_bill", y="tip", hue="time", data=tips)
plt.title("Regression Plot: Total Bill vs Tip (Faceted by Time)")
plt.show()
```
